## Comments
At $\alpha$ = 5000, bistability is pretty clear. Depending on how the system being started, it locks into one of two very stable states - either around 5000 P1 molecules with basically no P2, or the opposite. The initial conditions completely determine which state being end up with, which is exactly what we'd expect from a proper bistable switch. However, a mixed state where both are around 2500 molecules each were observed when initial values set to [0,0].

At $\alpha$ = 5, the whole system is just noisy chaos. The molecule counts are bouncing around randomly between roughly 1-4 molecules, and there's on hint of the two distinct states we saw before. The system can't maintain stable states because of random fluctuations.

The middle values ($\alpha$ = 50 and 500) are where things get interesting. At $\alpha$ = 50, you can start to see some trends of bistability - the different initial conditions do lead to somewhat different behaviors - but there's still a lot of noise muddying the waters. By $\alpha$ = 500, the picture becomes much clearer. The system get distinct states emerging: one where P1 dominates (~500 molecules), one where P2 dominates (~500 molecules), and a mixed state where both are around 250 molecules each with u0=[0,0].
